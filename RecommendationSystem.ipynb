```python
import pandas as pd
import numpy as  np
import ast 
import nltk
from nltk.stem.porter import PorterStemmer 
ps = PorterStemmer()
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import cosine_distances
```

```python
credits = pd.read_csv("credits.csv")
movies = pd.read_csv("movies.csv")
```

```python
movies
```

```python
credits
```

```python
movies = movies.merge(credits  , on = "title")
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies = movies[["movie_id","title","overview","genres","keywords","cast","crew"]]
```

```python
movies
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.info()
```

```python
movies.duplicated().sum()
```

```python
def  convert(object):
    L = []
    for i in ast.literal_eval(object):
        L.append(i["name"])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
def  convert2(object):
    L = []
    count = 0
    for i in ast.literal_eval(object):
        if count != 5:
            L.append(i["name"])
            count += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
def  convert3(object):
    L = []
    for i in ast.literal_eval(object):
        if i['job'] == 'Director':
            L.append(i["name"])
    return L
```

```python
movies['crew'] = movies['crew'].apply(convert3)
```

```python
movies.head()
```

```python
movies['overview'] = movies["overview"].apply(lambda x: x.split())
```

```python
movies
```

```python
movies["tags"]= movies["overview"] + movies["cast"] + movies["crew"] + movies["genres"] + movies["keywords"]
movies
```

```python
df = movies[["movie_id", "title", "tags"]]
```

```python
df
```

```python
df["tags"] =df["tags"].apply(lambda x: " ".join(x))
```

```python
df
```

```python
df["tags"] =df["tags"].apply(lambda x:x.lower())
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv  = CountVectorizer(max_features= 500, stop_words="english")
```

```python
cv.fit_transform(df["tags"]).toarray().shape
```

```python
vectors = cv.fit_transform(df["tags"]).toarray()
```

```python
vectors
```

```python
def stem(text):
    x = []
    for i in text.split():
        x.append(ps.stem(i))
    return " ".join(x)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
df
```

```python
sim = cosine_similarity(vectors)
```

```python
sim
```

```python
sim.shape
```

```python
sorted(list(enumerate(sim[0])), reverse= True, key= lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    distance = sim[movie_index]
    movie_list= sorted(list(enumerate(distance)), reverse= True, key= lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(df.iloc[i[0]].title)
    
```

```python
recommend("Avatar")
```
